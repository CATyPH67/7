график количества претензий и эжемесечного платежа

```python
import matplotlib.pyplot as plt

with open('data2.csv', 'r') as file:
    data = file.readlines()
del data[0]
data = [[float(n) for n in x.split(',')] for x in data]

x = []
y = []
for i in range(len(data)):
    y.append(data[i][0])
    x.append(data[i][1])
    
colors = ['tab:blue', 'tab:cyan', 'tab:grey', 'tab:green', 'tab:purple', 'tab:orange', 'tab:red'] 
col = []

for i in range(len(y)):
    index = int(y[i] // 20)
    col.append(colors[index])

plt.xlabel("ежемесячный платеж")
plt.ylabel("кол-во претензий")
plt.scatter(x,y,  c = col)
plt.show()
```

гистограмма кол-ва претензий, используя subplots

```python
fig, ax = plt.subplots()
plt.hist(y, bins = range(0, 150, 20))
ax.xaxis.set_major_locator(plt.IndexLocator(base = 20, offset = 0))
plt.ylabel("всего кол-во претензий")
plt.show()
```

столбчатая гистограмма кол-ва претензий, используя bar

```python
count = [0, 0, 0, 0, 0, 0, 0]
for i in range(len(y)):
    countIndex = int(y[i] // 20)
    count[countIndex] += 1

labels = ['0-20', '20-40', '40-60', '60-80', '80-100', '100-120', '120-140'] 
colors = ['tab:blue', 'tab:cyan', 'tab:grey', 'tab:green', 'tab:purple', 'tab:orange', 'tab:red'] 

plt.bar(labels, count, color = colors)
plt.ylabel("всего кол-во претензий")
plt.show()
```

круговая диаграмма кол-ва претензий

```python
fig, ax = plt.subplots() 
ax.pie(count, labels = labels, colors = colors) 
plt.show() 
```
